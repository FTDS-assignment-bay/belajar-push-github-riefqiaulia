# caranya

1. copy ssh `link`
2. **git clone** `ssh link` -> output nya akan membuat folder baru
3. cd ke new folder tersebut
4. **git status** -> outputnya akan muncul nama-nama file di folder baru tsb
5. *git add .* -> artinya saya akan ngepush seluruh file di folder baru tsb

***bisa juga git add <nama file>*** -> untuk ngepush file tertentu

6. **git status** -> outputnya akan muncul nama nama file di folder baru yang sudah di add berwarna hijau
7. **git commit** -m "revisi_1"
8. `git push`
